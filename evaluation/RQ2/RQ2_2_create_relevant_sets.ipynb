```python
import numpy as np
# https://stackoverflow.com/questions/7825956/find-cities-within-given-mile-radius-from-address

df1 = df[df['state_name']=='Tennessee']
df1 = df1[['city', 'lat', 'lng']]
for town in towns:
    first_row = df1.loc[df1['city'] == town].iloc[0]
    town_lat = first_row['lat']
    town_lng = first_row['lng']
    print(town_lat)
    print(town_lng)
    # exctract all towns that are within 3 mile radius
    df1[town] = np.where(((df1.lat-town_lat)**2+(df1.lng-town_lng)**2 < 0.1), 'close', 'not')
    
df1
```

```python
# form a df of those towns
condition = (df1 == 'close').any(axis=1)
df2 = df[df['state_name']=='Tennessee']
result = df2[condition]
result.to_csv('../relevant_db/TN_historical_sites.csv')
```

```python
prompt = "Coming to Arkansas. I’m interested in visiting it’s state parks. \
    Can you please recommend to me cities or towns to stay in that are close to good state parks?"

# https://koordinates.com/layer/102903-arkansas-state-park-state-park-locations/
file_path = '../code/State_Parks_AR.csv'
df3 = pd.read_csv(file_path)

df1 = df[df['state_name']=='Arkansas']

for index, row in df3.iterrows():
    lat = row['lat']
    lng = row['lng']
    df1[index] = np.where(((df1.lat-lat)**2+(df1.lng-lng)**2 < 0.05), 'close', 'not')
    
df1
```

```python

# form a df of those towns
condition = (df1 == 'close').any(axis=1)
df2 = df[df['state_name']=='Arkansas']
result = df2[condition]
result
result.to_csv('../relevant_db/AR_state_parks.csv')
```
