```python
# single constraint vs generic

#create a pandas dataframe
import pandas as pd
path = '/content/all_models_generic.csv'

df = pd.read_csv(path)
df_g = df.drop(['Unnamed: 0'], axis=1)
#df = df.reset_index()
df_g
```

```python
#create a pandas dataframe
import pandas as pd
path='/content/all_models_single_constraint.csv'
df = pd.read_csv(path)
df_sc = df.drop(['Unnamed: 0'], axis=1)
#df = df.reset_index()
df_sc
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from collections import defaultdict
from itertools import combinations

all_taus = []
all_sys = []
prompts = df_g.prompt.unique()

for system in ['gemma', 'gpt35', 'claude35', 'gpt4o', 'llama31', 'mistral']:

    df_sys = df_g[df_g['model'] == system]  # similarity within the same system

    for prompt in prompts:
        # Dataframe with all entries for a specific query
        df_prmp = df_sys[df_sys['prompt'] == prompt]  # similarity within the same queries

        reasons_list = []  # This will store all the combined reasons per row for BLEU comparison
        for row in range(len(df_prmp.index)):
            reasons_per_row = ''
            for rec in ['rec_reasons1', 'rec_reasons2', 'rec_reasons3', 'rec_reasons4', 'rec_reasons5']:
                reason = df_prmp.iloc[row][rec]
                reason = reason.replace('\\n', ' ').lower()  # preprocess text (remove newlines and lowercase)
                reasons_per_row += reason
            reasons_list.append(reasons_per_row)

        # Dictionary to store BLEU scores for each row combination
        d_queries = defaultdict(float)

        # Calculate BLEU score for each pair of reasons in the dataframe
        for (n_1, n_2) in combinations(range(len(df_prmp.index)), 2):
            reference = reasons_list[n_1].split()  # Reference text as list of words
            candidate = reasons_list[n_2].split()  # Candidate text as list of words

            # SmoothingFunction
            smooth_fn = SmoothingFunction().method1

            # Compute BLEU score
            bleu_score = sentence_bleu([reference], candidate, smoothing_function=smooth_fn)

            d_queries[n_1] += bleu_score

        for key in d_queries.keys():
            # Build list of scores for the plot
            all_taus.append((d_queries[key] / (len(df_prmp.index) - 1)))
            # Build list of systems for the plot
            all_sys.append(system)
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from collections import defaultdict
from itertools import combinations

all_taus1 = []
all_sys1 = []
prompts = df_sc.prompt.unique()

for system in ['gemma', 'gpt35', 'claude35', 'gpt4o', 'llama31', 'mistral']:

    df_sys = df_sc[df_sc['model'] == system]  # Similarity within the same system

    for prompt in prompts:
        # Dataframe with all entries for a specific query
        df_prmp = df_sys[df_sys['prompt'] == prompt]  # Similarity within same queries

        reasons_list = []  # This will store all the combined reasons per row for BLEU comparison
        for row in range(len(df_prmp.index)):
            reasons_per_row = ''
            for rec in ['rec_reasons1', 'rec_reasons2', 'rec_reasons3', 'rec_reasons4', 'rec_reasons5']:
                reason = df_prmp.iloc[row][rec]
                reason = reason.replace('\\n', ' ').lower()  # Preprocess text (remove newlines and lowercase)
                reasons_per_row += reason
            reasons_list.append(reasons_per_row)

        d_queries = defaultdict(float)

        # Calculate BLEU score for each pair of reasons in the dataframe
        for (n_1, n_2) in combinations(range(len(df_prmp.index)), 2):
            reference = reasons_list[n_1].split()  # Reference text as list of words
            candidate = reasons_list[n_2].split()  # Candidate text as list of words

            # SmoothingFunction
            smooth_fn = SmoothingFunction().method1

            # Compute BLEU score
            bleu_score = sentence_bleu([reference], candidate, smoothing_function=smooth_fn)

            d_queries[n_1] += bleu_score
            d_queries[n_2] += bleu_score

        for key in d_queries.keys():
            # Build list of scores for the plot
            all_taus1.append(d_queries[key] / (len(df_prmp.index) - 1))
            # Build list of systems for the plot
            all_sys1.append(system)
```

```python
Bleu = pd.DataFrame({"system": all_sys+all_sys1, "Blue-similarity": all_taus+all_taus1, "condition": ['generic']*len(all_sys)+['single-constraint']*len(all_sys1)})
Bleu.to_csv("internal_blue.csv")
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from collections import defaultdict
from itertools import combinations

all_taus = []
all_sys = []
prompts = df_g.prompt.unique()

def def_empty_string():
    return ""

for prompt in prompts:  # per prompt
    res_llm = defaultdict(def_empty_string)
    df_prmp = df_g[df_g['prompt'] == prompt]

    for system in ['gemma', 'gpt35', 'claude35', 'gpt4o', 'llama31', 'mistral']:  # per system
        df_sys = df_prmp[df_prmp['model'] == system]
        res_docs = ''  # a combined document of responses for the same prompt
        for row in range(len(df_sys.index)):
            res_doc = ''
            for rec in ['rec_reasons1', 'rec_reasons2', 'rec_reasons3', 'rec_reasons4', 'rec_reasons5']:
                reason = df_sys.iloc[row][rec]
                reason = reason.replace('\\n', '')
                # preprocess text (remove punctuation and lowercase)
                reason = reason.lower()
                # concatenate the reasons
                res_doc += reason  # a response
            res_docs += res_doc
        res_llm[system] = res_docs  # store concatenated reasons per system

    # Compare each system's response with every other system using BLEU score
    for (sys1, sys2) in combinations(['gemma', 'gpt35', 'claude35', 'gpt4o', 'llama31', 'mistral'], 2):
        reference = res_llm[sys1].split()  # reference system's response as a list of words
        candidate = res_llm[sys2].split()  # candidate system's response as a list of words

        # Smoothing function
        smooth_fn = SmoothingFunction().method1

        # Compute BLEU score
        bleu_score = sentence_bleu([reference], candidate, smoothing_function=smooth_fn)

        print(f'The BLEU score between {sys1} and {sys2} is {bleu_score}')

        # Append the results for both systems
        all_sys.append(sys1)
        all_taus.append(bleu_score)
        all_sys.append(sys2)
        all_taus.append(bleu_score)
```
